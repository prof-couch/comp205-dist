# The Visual Display of Quantitative Information

# Semiology of Graphics
* Jacques Bertin, *Semiology of Graphics* was a work before Tufte. 
* *Semiology* is the study of the language of symbols.
* In this work, which is Bertin's Ph.D. thesis, several really basic principles are developed. 

# Bertin's principles of data visualization. 
## 1. Data visualization has several *axes*. 

Objects depicting data can vary in: 
  
  * **Size**: how large is the object? 
  * **Color**: what color is the object? 
  * **Brightness**: How bright is the object? 
  * **Shape**: What shape is the object? 
  * **Position**: Where is the object on the page? 
  * **Texture**: What is the pattern on the object? 
  
## 2. These axes have different qualities in representing information. 
Bertin -- a cartographer -- complains about the misuse of symbols on maps. He concludes that there are several different axes of representation: 
  
  * **magnitude**: the value of a datum. 
  * **order**: the relative size of data. 
  * **distinction**: the fact that two visual objects represent different things. Equivalently: class membership or non-membership. 
  * **similarity**: the fact that two visual objects represent similar data. 
  
## 3. This leads to rules of data representation: 
Based upon the properties of human perception, 
  
  * **magnitude** can only be represented by size or position. 
  * **order** can be represented by anything that represents size, but also by brightness. 
  * **distinction** can be represented by anything that represents order, and also by color.
  * **similarity** can be represented by anything that represents distinction, and also by texture and shape. 
  
* Thus, there is a hierarchy of graphical attributes that represents different aspects of a possibly multi-dimensional space. 
  
  * size or position are most expressive. 
  * followed by brightness, 
  * followed by color, 
  * followed by texture and shape. 

## Thus, the map-maker's task is to: 
* Utilize the main three axes of x-y position and size to represent magnitude.  
* Use brightness to depict order. 
* Use color to depict distinction. 

# Silently and without attribution, all modern maps obey these principles! 
The typical iconography/semiology of a map includes that: 
* x,y depict latitude and longitude. 
* size represents population. 
* color, brightness represent some other categories.
* Texture is unused or represents similar areas on a map, e.g., what are the city limits? 

# The significance of Bertin's rules
* You can of course break these rules in representing data. 
* You cannot expect the average untrained human being to be able to percieve relationships between misused representation axes.
* The absolute worst thing you can do -- according to Bertin -- is to use texture and expect humans to be able to perceive magnitude of the data. 
* There are smaller sins also, including expecting humans to infer ordering from color. We have an innate ability to infer ordering from brightness, but no innate map from color to order. 
* Such interpretation can be learned by training, but is not innate. 

# Tufte's principles
* After Bertin, Tufte refined the principles. 
* **Chartjunk** is unnecessary detail in a depiction. 
* Misrepresentational practices that Tufte criticizes: 

  * "Three-dimensional graphics" that artificially enlarge depictions. 
  * Making data proportional to radius instead of area, thus making variation seem larger than it is.
