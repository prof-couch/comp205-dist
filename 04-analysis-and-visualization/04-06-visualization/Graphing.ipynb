### Line graphs 

We would typically start a linegraph the same way (by importing), but we have already done this so we can go right into making our data set.

We will first create a set of times, then use these times to find a value for a function or functions.

If we want to display multiple lines on the same plot we can do this as well, and even use different markers or colors for each of the lines.

```python
t = np.arange(0., 7., 0.1)

plt.plot(t,np.cos(t)*60)

plt.plot(t, t, 'o--', t, t**3+50, 'g^')
```

### More graphs:
We can set up a function that will take our t values and return respective f(t) values, then plot those as well

Futhermore, we can add a legend, a grid, a scale, and even add annotations on the graph.

```python
def x(t):
    return np.exp(t) * np.cos(.5*np.pi*t)
plt.plot(t,x(t))

plt.legend(['e^t + cos(.5(pi)(t))'])

plt.grid()

plt.xlim(1,10)

plt.annotate('local max', xy=(4.3, 67), xytext=(8, -200),
            arrowprops=dict(facecolor='black', shrink=0.05),
            )

plt.show()
```

#### Plotting a curve:
This code is from  'Getting Started with Plotly and Jupyter!' Youtube Video 

```python
import plotly.plotly as py
import plotly.graph_objs as go
import ipywidgets as widgets
import numpy as np
from scipy import special


x = np.linspace(0, np.pi, 1000)

layout = go.Layout(
    title = 'Simple Example',
    yaxis = dict(
        title = 'volts'
    ),
    xaxis = dict(
        title = 'nanoseconds'
    )
)

trace1 = go.Scatter(
    x=x,
    y=np.sin(x),
    mode = 'lines',
    line = dict(
        shape = 'spline'
    )
)

fig = go.Figure(data = [trace1], layout = layout)
py.iplot(fig,filename='Sin-Example')
```

#### On the plotly website you can see all the different types of graphs people have made, but I'll do a few more.

```python
t = np.arange(0., 7., 0.1)


layout = go.Layout(
    title = 'Same curve from before',
    yaxis = dict(
        title = 'Y value'
    ),
    xaxis = dict(
        title = 'X value'
    )
)
def x(t):
    return np.exp(t) * np.cos(.5*np.pi*t)
trace1 = go.Scatter(
    x=t,
    y=x(t),
    mode = 'lines',
    line = dict(
        shape = 'spline'
    )
)

fig = go.Figure(data = [trace1], layout = layout)
py.iplot(fig,filename='Exponential-Cosine')
```

## Importing Data:
Plotly works very well with dataframes, so we will import the data as a dataframe.
To make sure everything went well, we will use head() to visualalize the data.
Make sure to use the header delimeter to mark the header row, and the nrows delimeter so that we don't have to import all of the data, let's just say 30 rows of data.

```python
import pandas
from plotly.tools import FigureFactory as FF


df = pandas.read_csv('NYC_weather.csv')

sample_data_table = FF.create_table(df.head())
py.iplot(sample_data_table, filename='sample-data-table')


tracea = go.Scatter(x=df['DATE'], y=df['TMAX'], mode='lines', name='TMAX' )
traceb = go.Scatter(x=df['DATE'], y=df['TMIN'], mode='lines', name='TMIN')

layout = go.Layout(title='Simple Plot from csv data',
                   plot_bgcolor='rgb(230, 230,230)')

fig = go.Figure(data=[tracea, traceb], layout=layout)

py.iplot(fig, filename='simple-plot-from-csv')
```
